# Hello Manim!

# Shapes

```python
%%manim -qm -v WARNING Shapes

class Shapes(Scene):
    def construct(self):
        ######Code######
        #Making shapes
        circle = Circle()
        square = Square()
        triangle=Triangle()

        #Showing shapes
        self.play(Create(circle))
        self.wait(3)
        self.play(FadeOut(circle))
        self.wait(3)
        self.play(GrowFromCenter(square))
        self.wait(3)
        self.play(Transform(square,triangle))
        self.wait(3)
```

# Texts

```python
%%manim -qm -v WARNING makeText
class makeText(Scene):
    def construct(self):
        #######Code#######
        #Making text
        first_line = Text("Manim is fun")
        second_line = Text("and useful")
        final_line = Text("Hope you like it too!", color=BLUE)
        color_final_line = Text("Hope you like it too!")

        #Coloring
        color_final_line.set_color_by_gradient(BLUE,PURPLE)

        #Position text
        second_line.next_to(first_line, DOWN)

        #Showing text
        self.wait(1)
        self.play(Write(first_line), Write(second_line))
        self.wait(1)
        self.play(FadeOut(second_line), ReplacementTransform(first_line, final_line))
        self.wait(1)
        self.play(Transform(final_line, color_final_line))
        self.wait(2)
```

```python
%%manim -qm -v WARNING Equations

class Equations(Scene):
    def construct(self):
        # Equation 1 (full LaTeX)
        first_eq = MathTex(r"J(\theta) = -\frac{1}{m} \left[\sum_{i=1}^{m} y^{(i)} \log{h_{\theta}(x^{(i)})} + (1-y^{(i)}) \log{(1-h_{\theta}(x^{(i)}))} \right]")

        # Equation 2 (split components)
        second_eq_parts = [
            r"J(\theta_{0}, \theta_{1})", "=", r"\frac{1}{2m}", r"\sum\limits_{i=1}^m", "(",
            r"h_{\theta}(x^{(i)})", "-", r"y^{(i)}", ")^2"
        ]
        second_eq = MathTex(*second_eq_parts)

        # Color equation parts
        for i, part in enumerate(second_eq):
            if i < 2:
                part.set_color(PURPLE)
            else:
                part.set_color("#00FFFF")

        # Descriptions
        des1 = Text("You can write any math equations like this...")
        des2 = Text("Or this...")
        des3 = Text("And it looks nice!!")

        # Positioning
        des1.shift(2 * UP)
        des2.shift(2 * UP)

        # Animations
        self.play(Write(des1))
        self.play(Write(first_eq))
        self.play(ReplacementTransform(des1, des2), Transform(first_eq, second_eq))
        self.wait(1)

        self.add(second_eq)
        self.wait(1)

        self.play(
            AnimationGroup(
                second_eq.animate.shift(DOWN),
                FadeOut(second_eq),
                lag_ratio=0.0
            ),
            AnimationGroup(
                first_eq.animate.shift(DOWN),
                FadeOut(first_eq),
                lag_ratio=0.0
            ),
            Transform(des2, des3)
        )
        self.wait(2)
```

# 2D Coordinate Scene

```python
%%manim -qm -v WARNING Tute1

class Tute1(Scene):
    def construct(self):
        plane = NumberPlane(x_range=[-7, 7, 1], y_range=[-4, 4, 1]).add_coordinates()
        box = Rectangle(stroke_color=GREEN_C, stroke_opacity=0.7, fill_color=RED_B, fill_opacity=0.5, height=1, width=1)

        dot = always_redraw(lambda: Dot().move_to(box.get_center()))

        self.play(FadeIn(plane), run_time=6)
        self.wait()
        self.add(box, dot)
        self.play(box.animate.shift(RIGHT * 2), run_time=4)
        self.wait()
        self.play(box.animate.shift(UP * 3), run_time=4)
        self.wait()
        self.play(box.animate.shift(DOWN * 5 + LEFT * 5), run_time=4)
        self.wait()
        self.play(box.animate.shift(UP * 1.5 + RIGHT * 1), run_time=4)
        self.wait()
```

```python
%%manim -qm -v WARNING Graphing
import math

class Graphing(Scene):
    def construct(self):
        # Axes setup
        axes = Axes(
            x_range=[-5, 5],
            y_range=[-4, 4],
            x_length=10,
            y_length=6,
            axis_config={"color": BLUE}
        )
        axes.add_coordinates()

        # Define functions
        def func1(x):
            return x**2

        def func2(x):
            return x**3

        # Plot graphs
        graph1 = axes.plot(func1, color=WHITE)
        graph_label1 = axes.get_graph_label(graph1, label=MathTex("x^2"))

        graph2 = axes.plot(func2, color=WHITE)
        graph_label2 = axes.get_graph_label(graph2, label=MathTex("x^3"))

        # Vertical and horizontal lines to point (1, 1^2) = (1, 1)
        x_val = 1
        y_val = func1(x_val)
        vert_line = axes.get_vertical_line(axes.input_to_graph_point(x_val, graph1), color=YELLOW)
        horz_line = axes.get_horizontal_line(axes.input_to_graph_point(x_val, graph1), color=YELLOW)

        # Dot at (1, 1)
        point = Dot(axes.coords_to_point(x_val, y_val))

        # Animations
        self.play(Create(axes), Create(graph1), Write(graph_label1))
        self.wait(1)
        self.play(Create(vert_line), Create(horz_line))
        self.add(point)
        self.wait(1)
        self.play(
            Transform(graph1, graph2),
            Transform(graph_label1, graph_label2)
        )
        self.wait(2)
```

## Lorenz Attractor

```python
%%manim -qm -v WARNING LorenzAttractor2D

import numpy as np
from scipy.integrate import odeint
from manim import *

class LorenzAttractor2D(Scene):
    def construct(self):
        # Create a 2D axes
        axes = Axes(
            x_range=[-20, 20, 5],
            y_range=[-30, 30, 5],
            x_length=10,
            y_length=6,
            tips=False
        )
        self.add(axes)

        # Time values and initial conditions
        t_vals = np.linspace(0, 40, 10000)
        initial = [0, 1, 1.05]
        sigma, beta, rho = 10, 8/3, 28

        # Lorenz differential equations
        def lorenz_deriv(pos, t):
            x, y, z = pos
            dx = sigma * (y - x)
            dy = x * (rho - z) - y
            dz = x * y - beta * z
            return [dx, dy, dz]

        # Solve ODE
        points = odeint(lorenz_deriv, initial, t_vals)

        # Project to 2D (x, y)
        coords = [axes.c2p(x, y) for x, y, _ in points]

        # Create path
        path = VMobject()
        path.set_points_as_corners(coords)
        path.set_stroke(WHITE, width=2)

        self.play(Create(path), run_time=6)
        self.wait()
```

## Riemann Summation

```python
%%manim -qm -v WARNING RiemannIntegralModern

import numpy as np

class RiemannIntegralModern(Scene):
    def construct(self):
        # Setup axes
        axes = Axes(
            x_range=[-2, 6],
            y_range=[0, 4],
            x_length=8,
            y_length=4,
            axis_config={"include_numbers": True},
        ).to_edge(LEFT)
        self.play(Create(axes))

        # Define functions
        funcs = [
            (lambda x: 0.1 * (x - 2)**2 + 1, r"(x-2)^2 + 1"),
            (lambda x: (-1 * (x - 3)**3 + 3*(x - 3) + 3) / 15, r"-(x-3)^3 + 3(x-3) + 3"),
            (lambda x: (np.sin(x / 1.5) + 1) / 2, r"\sin\left(\frac{x}{1.5}\right) + 1"),
            (lambda x: x / 10 + 1, r"\frac{x}{10} + 1"),
        ]

        x_min, x_max = 0, 5

        def create_riemann_rects(f, dx):
            rects = VGroup()
            for i in np.arange(x_min, x_max, dx):
                height = f(i)
                rect = Rectangle(
                    width=dx * axes.x_axis.unit_size,
                    height=height * axes.y_axis.unit_size,
                    stroke_color=BLACK,
                    fill_color=interpolate_color(BLUE, YELLOW, (height - 0)/(4 - 0)),
                    fill_opacity=0.75
                )
                rect.move_to(axes.c2p(i + dx/2, height / 2))
                rects.add(rect)
            return rects

        # Animate through each function
        graph = None
        for f, latex_label in funcs:
            # Plot function
            new_graph = axes.plot(f, color=WHITE)
            graph_label = MathTex(latex_label).scale(1.2).to_edge(RIGHT)

            if graph:
                self.play(Transform(graph, new_graph), FadeOut(label), FadeIn(graph_label))
            else:
                self.play(Create(new_graph), FadeIn(graph_label))
            graph = new_graph
            label = graph_label

            # Initial rectangles
            rect_group = create_riemann_rects(f, 0.2)
            self.play(FadeIn(rect_group), run_time=1.5)

            # Animate rectangle refinement
            for dx in [0.15, 0.1, 0.05]:
                new_rects = create_riemann_rects(f, dx)
                self.play(Transform(rect_group, new_rects), run_time=2)

            self.play(FadeOut(rect_group), FadeOut(graph_label))

        self.play(FadeOut(graph), run_time=1.5)
```
